## using statistical methods

```python


# --- IQR Outlier Detection for Specific Columns with Count ---
def detect_outliers_iqr(df):
    outliers = pd.DataFrame()
    outlier_counts = {}

    # Specify the columns to check for outliers
    columns_to_check = ['Price per Unit', 'Units Sold', 'Total Sales', 'Operating Profit', 'Operating Margin']

    for column in columns_to_check:  
        if column in df.columns:
            Q1 = df[column].quantile(0.25)
            Q3 = df[column].quantile(0.75)
            IQR = Q3 - Q1

            lower_bound = Q1 - 1.5 * IQR
            upper_bound = Q3 + 1.5 * IQR

            # Detect outliers for the current column
            outliers_column = df[(df[column] < lower_bound) | (df[column] > upper_bound)]

            # Count the outliers
            outlier_counts[column] = len(outliers_column)

            if not outliers_column.empty:
                outliers = pd.concat([outliers, outliers_column])

    # Remove duplicates from the final outlier DataFrame
    outliers.drop_duplicates(inplace=True)

    print("Outliers detected using IQR:\n", outliers)
    print("\nOutlier Counts by Column:\n", outlier_counts)

# Example Usage
# df = pd.read_excel('path/to/your/excel_file.xlsx')  # Load your dataset
detect_outliers_iqr(df)
```

## normalization

```python
import pandas as pd
import numpy as np

# --- IQR Outlier Detection and Normalization for Specific Columns ---
def normalize_outliers_iqr(df):
    # Specify the columns to check for outliers
    columns_to_check = ['Price per Unit', 'Units Sold', 'Total Sales', 'Operating Profit', 'Operating Margin']
    
    # Loop through each column to detect and normalize outliers
    for column in columns_to_check:  
        if column in df.columns:
            # Calculate the IQR for the column
            Q1 = df[column].quantile(0.25)
            Q3 = df[column].quantile(0.75)
            IQR = Q3 - Q1

            # Calculate the lower and upper bounds
            lower_bound = Q1 - 1.5 * IQR
            upper_bound = Q3 + 1.5 * IQR

            # Normalize outliers by replacing with bounds
            df[column] = np.where(df[column] < lower_bound, lower_bound, df[column])
            df[column] = np.where(df[column] > upper_bound, upper_bound, df[column])

            print(f"Outliers in column '{column}' have been normalized.")
    
    return df

# Example Usage
# df = pd.read_excel('path/to/your/excel_file.xlsx')  # Load your dataset
df_normalized = normalize_outliers_iqr(df)

# Display the normalized DataFrame
print(df_normalized.head())
```

```python
import pandas as pd
import numpy as np

# --- IQR Outlier Detection for Specific Columns After Normalization ---
def detect_outliers_iqr_after_normalization(df):
    outlier_counts = {}
    
    # Specify the columns to check for outliers
    columns_to_check = ['Price per Unit', 'Units Sold', 'Total Sales', 'Operating Profit', 'Operating Margin']
    
    # Loop through each column to check for outliers
    for column in columns_to_check:  
        if column in df.columns:
            Q1 = df[column].quantile(0.25)
            Q3 = df[column].quantile(0.75)
            IQR = Q3 - Q1

            lower_bound = Q1 - 1.5 * IQR
            upper_bound = Q3 + 1.5 * IQR

            # Detect outliers for the current column after normalization
            outliers_column = df[(df[column] < lower_bound) | (df[column] > upper_bound)]

            # Count the outliers
            outlier_counts[column] = len(outliers_column)

    # Display outlier counts after normalization
    print("\nOutlier Counts After Normalization by Column:\n", outlier_counts)

# Example Usage
# df = pd.read_excel('path/to/your/excel_file.xlsx')  # Load your dataset
detect_outliers_iqr_after_normalization(df_normalized)
```
